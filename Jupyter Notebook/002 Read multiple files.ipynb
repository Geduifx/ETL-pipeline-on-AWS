```python
import boto3 # Python integration with AWS.
import pandas as pd # Read csv.
from io import StringIO # Read csv on AWS.
```

```python
# Variables for data on AWS S3.
s3 = boto3.resource('s3')
bucket = s3.Bucket('xetra-1234')
```

```python
# List of filtered CSVs.
# Two filters.
bucket_obj1 = bucket.objects.filter(Prefix='2022-02-15/')
bucket_obj2 = bucket.objects.filter(Prefix='2022-02-16/')
objects = [obj for obj in bucket_obj1] + [obj for obj in bucket_obj2]
```

```python
# Print list.
objects
```

```python
# Initial data frame.
csv_obj_init = bucket.Object(key=objects[0].key).get().get('Body').read().decode('utf-8')
data = StringIO(csv_obj_init)
df_init = pd.read_csv(data, delimiter=',')
```

```python
# Columns of initial data frame.
df_init.columns
```

```python
# Read csv.

# Initialize single data frame for all dfs.
df_all = pd.DataFrame(columns=df_init.columns)

# Loop for multiple csvs.
for obj in objects:
    csv_obj = bucket.Object(key=obj.key).get().get('Body').read().decode('utf-8')
    # Read csvs with StringIO (needed for AWS) and Pandas
    data = StringIO(csv_obj)
    df = pd.read_csv(data, delimiter=',')
    # Single data frame for all dfs.
    df_all = df_all.append(df, ignore_index=True)
```

```python
df_all
```

```python
# DF with filtered columns.
columns = ['ISIN', 'Date', 'Time', 'StartPrice', 'MaxPrice', 'MinPrice', 'EndPrice', 'TradedVolume']
df_all = df_all.loc[:, columns]
```

```python
df_all
```

```python
# Filter out empty lines.
df_all.dropna(inplace=True)
```
